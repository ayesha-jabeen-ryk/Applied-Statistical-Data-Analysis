```python
# Import libraries 

from collections import Counter
import re
```

# Frequency of Words

```python
file_path = 'textTask.txt' # Text is stored in text file.
```

```python
with open(file_path, 'r', encoding='utf-8') as file:
    content = file.read().lower()
```

```python
words = re.findall(r'\b\w+\b', content)
```

```python
total_words = len(words)
print(f"Total number of words (tokens): {total_words}")
```

```python
word_frequencies = Counter(words)
```

```python
print("Word frequencies in the file:\n")
for word, frequency in word_frequencies.items():
    print(f"{word}: {frequency}")
```

# Types of words

```python
import nltk
nltk.download('punkt')   # Tokenizer
nltk.download('averaged_perceptron_tagger')
```

```python
sentences = nltk.sent_tokenize(content)  # Split into sentences
words = [nltk.word_tokenize(sentence) for sentence in sentences]
```

```python
pos_tagged = [nltk.pos_tag(word) for word in words]
```

```python
print("POS Tagged words (word, part-of-speech tag):")
for sentence in pos_tagged:
    print(sentence)
```

# Dictionary of Words

```python
from collections import defaultdict
```

```python
word_dict = defaultdict(list)
```

```python
for sentence in pos_tagged:
    for word, pos in sentence:
        if pos.startswith('NN'):  # Nouns
            word_dict['nouns'].append(word)
        elif pos.startswith('VB'):  # Verbs
            word_dict['verbs'].append(word)
        elif pos.startswith('JJ'):  # Adjectives
            word_dict['adjectives'].append(word)
        elif pos.startswith('RB'):  # Adverbs (Optional)
            word_dict['adverbs'].append(word)

# Print the dictionary
print("Dictionary of categorized words:")
for word_type, words in word_dict.items():
    print(f"{word_type.capitalize()}: {words}")
```

# Words length

```python
words = nltk.word_tokenize(content)

# Step 3: Calculate word lengths
word_lengths = {word: len(word) for word in words}

# Step 4: Print the word lengths
print("Word lengths in the file:\n")
for word, length in word_lengths.items():
    print(f"{word}: {length}")
```
